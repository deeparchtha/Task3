```python
#Employee Performance Bonus Eligibility
# Input
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Find highest score
highest_score = max(employees.values())
# Find employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]
# Output
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
#Search Query Keyword Analysis
# Input
query = "Buy mobile phone buy phone online"
# Convert to lowercase
query = query.lower()
# Remove punctuation
import string
query = query.translate(str.maketrans("", "", string.punctuation))
# Split words
words = query.split()
# Count frequency
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
# Display words searched more than once
result = {word: count for word, count in frequency.items() if count > 1}
print(result)
```

```python
#Sensor Data Validation
# Input
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
# Store valid readings (even numbers)
valid_readings = []
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))
# Output
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Email Domain Usage Analysis
# Input
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
# Count domains
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
# Calculate percentage
total = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#Sales Spike Detection
# Input
sales = [1200, 1500, 900, 2200, 1400, 3000]
# Calculate average sales
average_sales = sum(sales) / len(sales)
# Detect days where sales are more than 30% above average
threshold = average_sales * 1.30
for index, value in enumerate(sales):
    if value > threshold:
        print(f"Day {index + 1}: {value}")
```

```python
#Duplicate User ID Detection
# Input
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Count occurrences
frequency = {}
for user in user_ids:
    frequency[user] = frequency.get(user, 0) + 1
# Display duplicates
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
